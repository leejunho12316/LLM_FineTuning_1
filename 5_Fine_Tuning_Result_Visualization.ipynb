```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.facecolor"] = "white"
plt.rcParams["axes.facecolor"]   = "white"
plt.rcParams["savefig.facecolor"] = "white"
plt.rcParams["text.color"]        = "#1a1a1a"
plt.rcParams["axes.labelcolor"]   = "#1a1a1a"
plt.rcParams["xtick.color"]       = "#1a1a1a"
plt.rcParams["ytick.color"]       = "#1a1a1a"

OUTPUT_DIR = "4.Visualization"
os.makedirs(OUTPUT_DIR, exist_ok=True)

KEYS    = ["importance", "department", "complaint_type", "emotion"]
KEYS_KO = ["중요도", "전달부서", "민원유형", "감정상태"]
MODELS  = ["Qwen2.5-0.5b", "Qwen2.5-1.5b", "Qwen2.5-3b", "Qwen2.5-7b"]

results = {
    "Qwen2.5-0.5b": {
        "base":      {"importance": 10.0, "department":  6.7, "complaint_type": 36.7, "emotion": 33.3},
        "finetuned": {"importance": 69.1, "department": 73.7, "complaint_type": 77.9, "emotion": 84.4},
    },
    "Qwen2.5-1.5b": {
        "base":      {"importance": 46.7, "department": 36.7, "complaint_type": 13.3, "emotion": 46.7},
        "finetuned": {"importance": 78.4, "department": 81.0, "complaint_type": 82.8, "emotion": 87.5},
    },
    "Qwen2.5-3b": {
        "base":      {"importance": 63.3, "department": 26.7, "complaint_type": 63.3, "emotion": 73.3},
        "finetuned": {"importance": 83.4, "department": 81.0, "complaint_type": 84.0, "emotion": 88.9},
    },
    "Qwen2.5-7b": {
        "base":      {"importance": 73.3, "department": 56.7, "complaint_type": 70.0, "emotion": 80.0},
        "finetuned": {"importance": 84.6, "department": 83.2, "complaint_type": 83.2, "emotion": 89.1},
    },
}

COLOR_BASE   = "#A8C8DC"
COLOR_FT     = "#2060A0"
COLOR_MODELS = ["#4C9BE8", "#E8844C", "#5CB85C", "#D45C8A"]

print("데이터 로드 완료:", list(results.keys()))
```

```python
# 그래프 1: 모델별 Fine-Tuning 전후 비교 (2×2 서브플롯)

fig, axes = plt.subplots(2, 2, figsize=(16, 11), sharey=True)
fig.suptitle("Qwen2.5 모델별 Fine-Tuning 전후 키워드 정답률", fontsize=17, fontweight="bold", y=1.01)
fig.patch.set_facecolor("white")

x = np.arange(len(KEYS))
width = 0.38

for ax, model in zip(axes.flatten(), MODELS):
    base_vals = [results[model]["base"][k]      for k in KEYS]
    ft_vals   = [results[model]["finetuned"][k]  for k in KEYS]

    b1 = ax.bar(x - width/2, base_vals, width, label="Base",       color=COLOR_BASE, edgecolor="white", linewidth=0.8)
    b2 = ax.bar(x + width/2, ft_vals,   width, label="Fine-Tuned", color=COLOR_FT,   edgecolor="white", linewidth=0.8)

    for bar in b1:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.8,
                f"{h:.1f}%", ha="center", va="bottom", fontsize=8, color="#444")
    for bar in b2:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.8,
                f"{h:.1f}%", ha="center", va="bottom", fontsize=8, fontweight="bold", color="#0a3f70")

    ax.set_title(model, fontsize=13, fontweight="bold", pad=8)
    ax.set_xticks(x)
    ax.set_xticklabels(KEYS_KO, fontsize=11)
    ax.set_ylim(0, 108)
    ax.set_ylabel("정답률 (%)", fontsize=10)
    ax.grid(axis="y", alpha=0.25, linestyle="--")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.legend(fontsize=9, loc="upper left")

plt.tight_layout()
path = f"{OUTPUT_DIR}/01_model_comparison.png"
plt.savefig(path, dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print(f"저장: {path}")
```

```python
# 그래프 2: 키워드별 전체 모델 비교 (2×2 서브플롯)

fig, axes = plt.subplots(2, 2, figsize=(16, 11))
fig.suptitle("키워드별 모델 크기에 따른 Fine-Tuning 전후 정답률", fontsize=17, fontweight="bold", y=1.01)

x = np.arange(len(MODELS))
width = 0.38

for ax, key, key_ko in zip(axes.flatten(), KEYS, KEYS_KO):
    base_vals = [results[m]["base"][key]      for m in MODELS]
    ft_vals   = [results[m]["finetuned"][key]  for m in MODELS]

    b1 = ax.bar(x - width/2, base_vals, width, label="Base",       color=COLOR_BASE, edgecolor="white")
    b2 = ax.bar(x + width/2, ft_vals,   width, label="Fine-Tuned", color=COLOR_FT,   edgecolor="white")

    for bar in b1:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.8,
                f"{h:.1f}%", ha="center", va="bottom", fontsize=8, color="#444")
    for bar in b2:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.8,
                f"{h:.1f}%", ha="center", va="bottom", fontsize=8, fontweight="bold", color="#0a3f70")

    ax.set_title(f"[{key_ko}] 키워드 정답률", fontsize=13, fontweight="bold", pad=8)
    ax.set_xticks(x)
    ax.set_xticklabels(MODELS, fontsize=10, rotation=10)
    ax.set_ylim(0, 108)
    ax.set_ylabel("정답률 (%)", fontsize=10)
    ax.grid(axis="y", alpha=0.25, linestyle="--")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.legend(fontsize=9, loc="upper left")

plt.tight_layout()
path = f"{OUTPUT_DIR}/02_keyword_comparison.png"
plt.savefig(path, dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print(f"저장: {path}")
```

```python
# 그래프 4: 모델 크기별 평균 정답률 추세

avg_base = [np.mean([results[m]["base"][k]      for k in KEYS]) for m in MODELS]
avg_ft   = [np.mean([results[m]["finetuned"][k]  for k in KEYS]) for m in MODELS]
MODEL_SIZES = [0.5, 1.5, 3, 7]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(MODEL_SIZES, avg_base, marker="o", markersize=9, linewidth=2.5,
        color=COLOR_BASE, markeredgecolor="#4a8aaa", label="Base")
ax.plot(MODEL_SIZES, avg_ft,   marker="s", markersize=9, linewidth=2.5,
        color=COLOR_FT,   markeredgecolor="#003a80", label="Fine-Tuned")
ax.fill_between(MODEL_SIZES, avg_base, avg_ft, alpha=0.1, color=COLOR_FT)

for xs, y1, y2 in zip(MODEL_SIZES, avg_base, avg_ft):
    ax.text(xs, y1 - 2.5, f"{y1:.1f}%", ha="center", fontsize=9.5, color="#4a8aaa")
    ax.text(xs, y2 + 1.2, f"{y2:.1f}%", ha="center", fontsize=9.5, fontweight="bold", color="#003a80")

ax.set_xticks(MODEL_SIZES)
ax.set_xticklabels([f"{s}B" for s in MODEL_SIZES], fontsize=12)
ax.set_xlabel("모델 파라미터 크기", fontsize=12)
ax.set_ylabel("평균 정답률 (%)", fontsize=12)
ax.set_ylim(0, 105)
ax.set_title("모델 크기별 Fine-Tuning 전후 평균 정답률 추세", fontsize=15, fontweight="bold", pad=12)
ax.grid(alpha=0.25, linestyle="--")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.legend(fontsize=11)

plt.tight_layout()
path = f"{OUTPUT_DIR}/04_model_size_trend.png"
plt.savefig(path, dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print(f"저장: {path}")
```

```python
# 그래프 5: 레이더 차트 (Base vs Fine-Tuned 전체 모델)

N = len(KEYS)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

fig, axes = plt.subplots(1, 2, figsize=(16, 7), subplot_kw={"projection": "polar"})
fig.suptitle("Fine-Tuning 전후 키워드 정답률 레이더 차트", fontsize=16, fontweight="bold")

titles = ["Base Model", "Fine-Tuned Model"]
modes  = ["base", "finetuned"]

for ax, title, mode in zip(axes, titles, modes):
    for model, color in zip(MODELS, COLOR_MODELS):
        vals = [results[model][mode][k] for k in KEYS]
        vals += vals[:1]
        ax.plot(angles, vals, linewidth=2, color=color, label=model)
        ax.fill(angles, vals, alpha=0.08, color=color)

    ax.set_thetagrids(np.degrees(angles[:-1]), KEYS_KO, fontsize=11)
    ax.set_ylim(0, 100)
    ax.set_rlabel_position(45)
    ax.set_yticks([20, 40, 60, 80, 100])
    ax.set_yticklabels(["20%", "40%", "60%", "80%", "100%"], fontsize=7.5, color="gray")
    ax.set_title(title, fontsize=13, fontweight="bold", pad=18)
    ax.legend(loc="upper right", bbox_to_anchor=(1.38, 1.15), fontsize=9)
    ax.grid(alpha=0.3)

plt.tight_layout()
path = f"{OUTPUT_DIR}/05_radar_chart.png"
plt.savefig(path, dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print(f"저장: {path}")
```
