# Rule-Based

```python
import json, re

def parsing(text : str) -> dict:
    match = re.search(r'\{.*?\}', text, re.DOTALL)
    if not match:
        return None
    raw = json.loads(match.group())
    return {k.lower(): v for k, v in raw.items()}

parsing_result = []
for i, row in df_original.iterrows():
    parsing_result.append(parsing(row['assistant']))

print(parsing_result)
```

```python
# 1. JSON : assistant가 JSON 형식을 만족해 parsing 가능한지
parsing_error_num = sum(1 for result in parsing_result if result is None)
print(f'파싱 실패 : {parsing_error_num} / {len(df_original)}')
```

```python
# 2. key : 키워드 네 가지를 다 포함하고 있는지
key_error_num = 0
required_keys = {'importance', 'department', 'complaint_type', 'emotion'}

for result in parsing_result:
    actual_keys = {key.lower() for key in result.keys()}

    if not required_keys.issubset(actual_keys):
          key_error_num += 1

print(f'키워드 실패 : {key_error_num} / {len(df_original)}')
```


3. value : 키워드 별 값이 올바른 값을 가지고 있는지

```python
value_map = {
    'importance' : ['높음','보통','낮음'],
    'department' : ['교통실', '기후환경본부', '주택실', '시민건강국', '문화본부', '경제실', '복지실', '여성가족실', '재난안전실', '분류 보류'],
    'complaint_type' : ['신고', '문의', '건의', '칭찬', '항의', '그 외'],
    'emotion' : ['긍정','중립','부정']
}

value_error_num = 0
for result in parsing_result:

    for key in ['importance', 'department', 'complaint_type', 'emotion']:
        if result[key].lower() not in value_map[key]:
            value_error_num += 1

print(f'값 실패 : {value_error_num} / {len(df_original)}')
```
