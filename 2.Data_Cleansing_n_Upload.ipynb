# 전처리

```python
#전처리 전
print(len(data))
```

```python
#1. Question 중복 제거
data.drop_duplicates(subset = ['Question'], inplace=True)
print(len(data))
```

```python
#2. Answer 결측치 제거
indexes = data[data['Answer'].isnull()].index
data.drop(indexes, inplace=True)
print(len(data))
```

```python
#3. 한성대 삭제
#2014년 11월 ~ 2015년 1월 title에 한성대 들어가는 민원 삭제.
indexes = data[data['title'].str.contains('한성대')].index
data.drop(indexes, inplace=True)
print(len(data))
```
